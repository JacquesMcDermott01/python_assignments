## 1.0 Research question: How do vary levels of infrastructure affect the concept of Minimal Differentiation? A simulation using the Hotelling Spacial Differentiation model

## 1.1 Introduction

#### The Hotelling model (found here https://www.jstor.org/stable/2224214) is one of the foundational models in spatial economics and industrial organization. Originally developed by Harold Hotelling in 1929, the model examines how firms choose locations when competing for consumers distributed across space. In its classic form, consumers are uniformly distributed along a one-dimensional axis and incur transportation costs when travelling to firms. It is often explained with the analogy of 2 identical ice cream vendors on a beach with equally distributed people sunbathing. Each ice cream seller cannot compete on quality but instead competes by strategically choose their locations in order to maximize profits . The model is particularly well known for demonstrating the principle of minimum differentiation, whereby the ice cream sellers choose to locate close to one another in equilibrium. This finding will be demonstrated later.

## 2.0 Hotelling's model - Definitions

#### Now we move on to deriving and solving the original model. We start with some definitions:

#### x = Distance
#### x_i = Location of the indifferent consumer
#### a = Location of shop A
#### b = Location of shop B
#### t = Transportation cost
#### U_a = Utility of consumers going to shop A
#### U_b = Utility of consumers going to shop B
#### D_A = Demand for shop A
#### D_B = Demand for shop B

## 2.1 Hotelling's model - Solving for location and price
$$
U_a = -p_A - t(x-x_A)
$$
$$
U_b = -p_B - t(x-x_B)
$$
#### For consumer i, the utility of going to either shop is the same, therefore:
$$
-p_A - t(x_i-x_A) = p_B + t(x_B-x_i)
$$
#### This solves for:
$$
x_i = 1/2(x_A + x_B + (p_B - p_A)/t)
$$
#### Now we add in demand functions. Every consumer left of the indifferent consumer will visit shop A, while every shop to the right of them will visit shop B. Therefore:
$$
    D_A = x_i ; 
    D_B = 1 - x_i
$$
$$
π_A = p_A . x_i ; π_B = p_B . (1-x_i)
$$
#### Both shops choose a price that maximises profit, therefore:
$$
π_A = p_A . [1/2x_A + 1/2x_B + (p_B - p_A)/2t)]
$$
$$
dπ_A/dp_A = 1/2x_A + 1/2x_B + p_B/t -2p_A/t = 0
$$
#### This leads to 2 simultanous equations
$$
(1) 2p_A = p_B + t(x_A + x_B)
$$
$$
(2) 2p_B = p_A + t(2 - x_A + x_B)
$$
#### Combining these and solving gives
$$
p_A = t/3(2+x_A+x_B)
$$
$$
p_B = t/3(4-x_A-x_B)
$$
#### Using the assumption of symmetrical locations, we know that both shops will be equally far away from the edge, therefore:
$$
x_A = y, 
x_B = 1-a
$$
$$
p_A = t/3(2+y+(y-1))
$$
$$
p_B = t/3(4-y-(1-y))
$$
$$
p_A = t; p_B = t
$$
$$
π_A = π_B = 1/2t
$$
#### These results are true for any value of y, meaning that any pair of symmetrical locations have the same profit. However, for a stable game both firms must be in the centre at 0.5 as either one would increase profits by moving to the centre. For example, say shop A is at 0.1 and shop B is at 0.9, but shop B breaks the symmetry by moving to location 0.6
$$
x_A = 0.1; x_B = 0.6
$$
$$
p_A = t/3(2+0.1+0.6)
$$
$$
p_A = 0.9t
$$
$$
p_B = t/3(4-0.1-0.6)
$$
$$
p_B = 1.1t
$$
$$
x_i = (0.1+0.6)/2 = 0.35
$$
$$
π_A = p_A*(x_i) = 0.9t*0.35 = 0.315t
$$
$$
π_B = p_B*(1-x_i) = 1.1t*(1-0.35) = 0.715t
$$
#### This is an unstable result as shop A will move closer to the centre to maximise profits. Shop B will do likewise until both are at location 0.5

## 3.0 Extension of the model

#### All of the answers so far have included t as a variable, now we want to elaborate on the transportation cost and see how it affects the model. Now transportation cost, t, depend on the following variables:
#### - F = Natural Friction. This represents factors like fuel cost and fatigue from walking, both of which affect the cost of travelling but are not affected by infrastructure
#### - R = Road quality, or Infrastructure quality. There is assumed to be 2 regions in this game, Left and Right. Region Left stretches between locations 0 and 0.5 while Region Right stretches between 0.5 and 1. The cost of travelling through Region Left will be affected by the quality of infrastructure in Region Left, represented by I_Left

#### Since this is complex addition, both firms will use a fixed price of 1 and therefore they can only compete by choosing a location. If no other changes were made, both firms would still choose the location 0.5. This is because all consumers must purchase 1 unit of the product so since both firms can ignore the transportation cost incurred by the consumers, they maximise access to consumers by choosing the location 0.5.

#### In order to prevent this, now consumers will no longer purchase if their utility is negative. We also assume all consumers value the good equally and that it gives them a utility of 2.

## 3.1 Extention definitions and equations

### Transportation cost is determined by Natural Friction divied by the Infrastructure Quality

$$
t_i(x) = F/R(x)

$$

U_i(x) = v - p - F/R(x) (x-x_i)

#### Customers only buy now if

## 4.2 Simulation - Defining Define consumer utility:

```python
def utility(x, firm_location, v=5, p=1, F=10):
    """
    Utility of a consumer at x from buying from a firm at firm_location.
    """
    t = transport_cost(x, F)
    distance = abs(x - firm_location)

    return v - p - t * distance
```

## 4.3 Simulation - Adding utility restriction

```python
def consumer_choice(x, a, b, v=5, p=1, F=10, utility_threshold=0):
    """
    Determines whether a consumer buys from Firm A, Firm B, or does not buy.

    Returns:
    - "A" if consumer buys from Firm A
    - "B" if consumer buys from Firm B
    - "None" if consumer does not buy
    """

    u_A = utility(x, a, v, p, F)
    u_B = utility(x, b, v, p, F)

    best_utility = max(u_A, u_B)

    if best_utility <= utility_threshold:
        return "None"

    if u_A > u_B:
        return "A"
    elif u_B > u_A:
        return "B"
    else:
        return "Tie"
```

## 4.4 Simulation - Simulate demand

```python
def compute_demand(a, b, v=5, p=1, F=10, utility_threshold=0, n_consumers=1000):
    """
    Computes demand for Firm A and Firm B by simulating consumers on [0,1].
    """

    consumers = np.linspace(0, 1, n_consumers)

    demand_A = 0
    demand_B = 0
    no_purchase = 0

    for x in consumers:
        choice = consumer_choice(x, a, b, v, p, F, utility_threshold)

        if choice == "A":
            demand_A += 1

        elif choice == "B":
            demand_B += 1

        elif choice == "Tie":
            demand_A += 0.5
            demand_B += 0.5

        else:
            no_purchase += 1

    demand_A = demand_A / n_consumers
    demand_B = demand_B / n_consumers
    no_purchase = no_purchase / n_consumers

    return demand_A, demand_B, no_purchase
```

## 4.5 Simulation - Compute profits

```python
def compute_profits(a, b, v=5, p=1, F=10, utility_threshold=2):
    """
    Computes profits for Firm A and Firm B.
    """

    demand_A, demand_B, no_purchase = compute_demand(
        a, b, v, p, F, utility_threshold
    )

    profit_A = p * demand_A
    profit_B = p * demand_B

    return profit_A, profit_B, demand_A, demand_B, no_purchase
```

## 4.6 Simulation - Compute best response functions for Firm A

```python
def best_response_A(
    b,
    locations,
    v=5,
    p=1,
    F=10,
    utility_threshold=2
):
    """
    Finds Firm A's profit-maximizing location given Firm B's location.
    """

    best_profit = -np.inf
    best_location = None

    for a in locations:

        profit_A, _, _, _, _ = compute_profits(
            a,
            b,
            v,
            p,
            F,
            utility_threshold
        )

        if profit_A > best_profit:

            best_profit = profit_A
            best_location = a

    return best_location, best_profit
```

## 4.7 Simulation - Compute best response functions for Firm B

```python
def best_response_B(
    a,
    locations,
    v=5,
    p=1,
    F=10,
    utility_threshold=2
):
    """
    Finds Firm B's profit-maximizing location given Firm A's location.
    """

    best_profit = -np.inf
    best_location = None

    for b in locations:

        _, profit_B, _, _, _ = compute_profits(
            a,
            b,
            v,
            p,
            F,
            utility_threshold
        )

        if profit_B > best_profit:

            best_profit = profit_B
            best_location = b

    return best_location, best_profit
```

```python
def solve_equilibrium(
    locations,
    v=5,
    p=1,
    F=10,
    utility_threshold=2,
    tolerance=1e-3,
    max_iterations=100
):
    """
    Solves for the Nash equilibrium locations.
    """

    # Initial guesses
    a = 0.25
    b = 0.75

    for iteration in range(max_iterations):

        old_a = a
        old_b = b

        # Firm A best responds
        a, _ = best_response_A(
            b,
            locations,
            v,
            p,
            F,
            utility_threshold
        )

        # Firm B best responds
        b, _ = best_response_B(
            a,
            locations,
            v,
            p,
            F,
            utility_threshold
        )

        # Check convergence
        if abs(a - old_a) < tolerance and abs(b - old_b) < tolerance:
            break

    return a, b
```

```python
# Step 1 - Define Infrastructure, Transportation cost & Utility function

def infrastructure(x, I_left=1, I_right=5):
    """Infrastructure quality at location x."""
    return I_left if x < 0.5 else I_right


def transport_cost(x, F=10, I_left=1, I_right=5):
    """Effective transport cost at consumer location x."""
    return F / infrastructure(x, I_left, I_right)


def utility(x, firm_location, v=3, p=1, F=10, I_left=1, I_right=5):
    """Consumer utility from buying from a firm at firm_location."""
    t = transport_cost(x, F, I_left, I_right)
    return v - p - t * abs(x - firm_location)

# Step 2 - Define consumer choice, demand function and profits

def consumer_choice(x, a, b, v=3, p=1, F=10,
                    I_left=1, I_right=5, utility_threshold=2):
    """Return the consumer's choice: 'A', 'B', 'Tie', or 'None'."""

    u_A = utility(x, a, v, p, F, I_left, I_right)
    u_B = utility(x, b, v, p, F, I_left, I_right)

    if max(u_A, u_B) <= utility_threshold:
        return "None"

    if u_A > u_B:
        return "A"
    elif u_B > u_A:
        return "B"
    else:
        return "Tie"


def compute_demand(a, b, v=3, p=1, F=10,
                   I_left=1, I_right=5, utility_threshold=2,
                   n_consumers=200):
    """Compute demand shares for both firms and the no-purchase share."""

    consumers = np.linspace(0, 1, n_consumers)

    demand_A = 0
    demand_B = 0
    no_purchase = 0

    for x in consumers:
        choice = consumer_choice(
            x, a, b, v, p, F, I_left, I_right, utility_threshold
        )

        if choice == "A":
            demand_A += 1
        elif choice == "B":
            demand_B += 1
        elif choice == "Tie":
            demand_A += 0.5
            demand_B += 0.5
        else:
            no_purchase += 1

    return (
        demand_A / n_consumers,
        demand_B / n_consumers,
        no_purchase / n_consumers
    )

def compute_profits(a, b, v=3, p=1, F=10,
                    I_left=1, I_right=5, utility_threshold=2,
                    n_consumers=200):
    """Compute profits and demand outcomes."""

    demand_A, demand_B, no_purchase = compute_demand(
        a, b, v, p, F, I_left, I_right,
        utility_threshold, n_consumers
    )

    profit_A = p * demand_A
    profit_B = p * demand_B

    return profit_A, profit_B, demand_A, demand_B, no_purchase

# Step 3 - Stimulate best responses

def best_response(firm, opponent_location, locations,
                  v=3, p=1, F=10,
                  I_left=1, I_right=5, utility_threshold=2,
                  n_consumers=200):
    """Find the best-response location for one firm."""

    best_profit = -np.inf
    best_location = None

    for loc in locations:

        if firm == "A":
            profit, _, _, _, _ = compute_profits(
                loc, opponent_location, v, p, F,
                I_left, I_right, utility_threshold, n_consumers
            )

        elif firm == "B":
            _, profit, _, _, _ = compute_profits(
                opponent_location, loc, v, p, F,
                I_left, I_right, utility_threshold, n_consumers
            )

        else:
            raise ValueError("firm must be either 'A' or 'B'")

        if profit > best_profit:
            best_profit = profit
            best_location = loc

    return best_location, best_profit

# Step 5 - Find Equilibrium

locations = np.linspace(0, 1, 101)

a_eq, b_eq = solve_equilibrium(
    locations,
    v=3,
    p=1,
    F=10,
    I_left=1,
    I_right=5,
    utility_threshold=2
)

_, _, no_purchase = compute_demand(
    a_eq, b_eq,
    v=3,
    p=1,
    F=10,
    I_left=1,
    I_right=5,
    utility_threshold=2
)

print("Equilibrium location A:", a_eq)
print("Equilibrium location B:", b_eq)
print("No-purchase share:", no_purchase)
```

```python
# Step 1 - Define Infrastructure, Transportation cost & Utility function

def infrastructure(x, I_left=0.5, I_right=10):
    """Infrastructure quality at location x."""
    return I_left if x < 0.5 else I_right


def transport_cost(x, F=10, I_left=0.5, I_right=10):
    """Effective transport cost at consumer location x."""
    return F / infrastructure(x, I_left, I_right)


def utility(x, firm_location, v=5, p=1, F=10, I_left=0.5, I_right=10):
    """Consumer utility from buying from a firm at firm_location."""
    t = transport_cost(x, F, I_left, I_right)
    return v - p - t * abs(x - firm_location)

# Step 2 - Define consumer choice, demand function and profits

def consumer_choice(x, a, b, v=5, p=1, F=10,
                    I_left=0.5, I_right=10, utility_threshold=2):
    """Return the consumer's choice: 'A', 'B', 'Tie', or 'None'."""

    u_A = utility(x, a, v, p, F, I_left, I_right)
    u_B = utility(x, b, v, p, F, I_left, I_right)

    if max(u_A, u_B) <= utility_threshold:
        return "None"

    if u_A > u_B:
        return "A"
    elif u_B > u_A:
        return "B"
    else:
        return "Tie"


def compute_demand(a, b, v=5, p=1, F=10,
                   I_left=0.5, I_right=10, utility_threshold=2,
                   n_consumers=200):
    """Compute demand shares for both firms and the no-purchase share."""

    consumers = np.linspace(0, 1, n_consumers)

    demand_A = 0
    demand_B = 0
    no_purchase = 0

    for x in consumers:
        choice = consumer_choice(
            x, a, b, v, p, F, I_left, I_right, utility_threshold
        )

        if choice == "A":
            demand_A += 1
        elif choice == "B":
            demand_B += 1
        elif choice == "Tie":
            demand_A += 0.5
            demand_B += 0.5
        else:
            no_purchase += 1

    return (
        demand_A / n_consumers,
        demand_B / n_consumers,
        no_purchase / n_consumers
    )

def compute_profits(a, b, v=5, p=1, F=10,
                    I_left=0.5, I_right=10, utility_threshold=2,
                    n_consumers=200):
    """Compute profits and demand outcomes."""

    demand_A, demand_B, no_purchase = compute_demand(
        a, b, v, p, F, I_left, I_right,
        utility_threshold, n_consumers
    )

    profit_A = p * demand_A
    profit_B = p * demand_B

    return profit_A, profit_B, demand_A, demand_B, no_purchase

# Step 3 - Stimulate best responses

def best_response(firm, opponent_location, locations,
                  v=5, p=1, F=10,
                  I_left=0.5, I_right=10, utility_threshold=2,
                  n_consumers=200):
    """Find the best-response location for one firm."""

    best_profit = -np.inf
    best_location = None

    for loc in locations:

        if firm == "A":
            profit, _, _, _, _ = compute_profits(
                loc, opponent_location, v, p, F,
                I_left, I_right, utility_threshold, n_consumers
            )

        elif firm == "B":
            _, profit, _, _, _ = compute_profits(
                opponent_location, loc, v, p, F,
                I_left, I_right, utility_threshold, n_consumers
            )

        else:
            raise ValueError("firm must be either 'A' or 'B'")

        if profit > best_profit:
            best_profit = profit
            best_location = loc

    return best_location, best_profit

# Step 5 - Find Equilibrium

def solve_equilibrium(locations,
                      v=5, p=1, F=10,
                      I_left=0.5, I_right=10,
                      utility_threshold=2,
                      n_consumers=200,
                      tolerance=1e-3,
                      max_iterations=100):
    """Solve for equilibrium locations using iterative best responses."""

    a = 0.25
    b = 0.75

    for iteration in range(max_iterations):

        old_a, old_b = a, b

        a, _ = best_response(
            "A", b, locations, v, p, F,
            I_left, I_right, utility_threshold, n_consumers
        )

        b, _ = best_response(
            "B", a, locations, v, p, F,
            I_left, I_right, utility_threshold, n_consumers
        )

        if abs(a - old_a) < tolerance and abs(b - old_b) < tolerance:
            break

    return a, b

# Step 6 - Find equilibrium
    
    locations = np.linspace(0, 1, 101)

a_eq, b_eq = solve_equilibrium(
    locations,
    v=5,
    p=1,
    F=10,
    I_left=0.5,
    I_right=10,
    utility_threshold=2
)

no_purchase = compute_profits(
    a_eq, b_eq,
    v=5,
    p=1,
    F=10,
    I_left=0.5,
    I_right=10,
    utility_threshold=2
)

print("Equilibrium location A:", a_eq)
print("Equilibrium location B:", b_eq)
print("No-purchase share:", no_purchase)

results = []

for I_right in np.linspace(1, 10, 20):

    a_eq, b_eq = solve_equilibrium(
        locations,
        v=5,
        p=1,
        F=10,
        I_left=0.5,
        I_right=I_right,
        utility_threshold=2
    )

     no_purchase = compute_profits(
        a_eq, b_eq,
        v=5,
        p=1,
        F=10,
        I_left=0.5,
        I_right=I_right,
        utility_threshold=2
    )

    results.append({
        "I_right": I_right,
        "a_eq": a_eq,
        "b_eq": b_eq,
        "no_purchase": no_purchase
    })

df = pd.DataFrame(results)

print(df.head())
```

## Conclusion

#### Contrary to my expectations, the results showed that firms may choose to be located in a region with poor infrastructure so that their access to consumers in those regions isn't cut off. This result will likely dimish if prices and per-unit costs were also simulated, as firms may find it more profitable to only service the demand of those in regions with good infrastructure. Still, it is an interesting result and can be instructive. There may be firms who operate low-profit businesses (such as supermarkets) who care more about access to as many customers as possible, who would want to be located in poor-infrastructure regions to maximise their customer base.
